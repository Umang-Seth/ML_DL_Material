```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("SaratogaHouses.csv")
```

```python
dataset
```

```python
dataset.head()
```

```python
X= dataset['livingArea'].values
y = dataset['price'].values
```

```python
y
```

```python
mean_x = np.mean(X)
mean_y = np.mean(y)
```

```python
mean_x, mean_y
```

```python
num = 0
den = 0

for i in range(len(X)):
    num += (X[i]- mean_x) * (y[i]-mean_y)
    den += (X[i]- mean_x) ** 2
    

slope = num/den
c = mean_y- mean_x*slope
```

```python
slope, c
```

```python
from scipy.stats import linregress
```

```python
linregress(X,y)
```

```python
prediction = []

for i in range(len(X)):
    y_pred= slope*X[i] + c
    prediction.append(y_pred)
```

```python
plt.scatter(X,y)
plt.plot(X,prediction, color = 'red')
```

```python
print(prediction[:10])
print(y[:10])
```

```python
numer=0
n = len(X)
for i in range(len(X)):
    numer += (prediction[i]-y[i]) ** 2
    
mse = numer / n

rmse = np.sqrt(mse)
```
